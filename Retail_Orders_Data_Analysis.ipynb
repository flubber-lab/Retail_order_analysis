```python
c1 = df[(df['Ship Mode'] == 'Second Class') & (df['Segment'] == 'Corporate')].shape[0]
c2 = df[(df['Ship Mode'] == 'Standard Class') & (df['Segment'] == 'Corporate')].shape[0]
c3 = df[(df['Ship Mode'] == 'First Class') & (df['Segment'] == 'Corporate')].shape[0]
print(c1,c2,c3)
```

```python
#Replacing NaN with most frequent value
df_clean = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
#Replacing 'unknown' with 'Not Available'
df_clean['Ship Mode'].replace('unknown','Not Available',inplace = True)
df_clean['Ship Mode'].unique()
```

```python
#Formatting date to DD/MM/YYY
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
df_clean['Order Date'] = df_clean['Order Date'].dt.strftime('%d/%m/%y')
df_clean
```

```python
df_clean.columns
```

```python
#Making SQL ready
df_clean.columns=df_clean.columns.str.lower().str.replace(" ","_")
```

```python
#Trimming white spaces using strip()
df_clean.rename(columns = lambda x: x.strip(), inplace = True)
```

```python
#Adding discount column with calculated value
discount = []
for i in range(len(df_clean)):
  discount.append((df_clean['list_price'][i])*(df_clean['discount_percent'][i]/100))
df_clean['discount'] = discount
```

```python
#Adding sale_price column with calculated values
df_clean['sale_price'] = (df_clean['list_price']-df_clean['discount']).round(2)
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
#Adding profit column with calculated values
df_clean['profit'] = (df_clean['sale_price']-df_clean['cost_price']).round(2)
```

```python
df_clean
```

```python
df_sales = df_clean.copy()
df_sales
```

```python
df_sales.drop(['ship_mode', 'segment', 'country', 'city',
       'state', 'postal_code', 'region'],axis=1,inplace=True)
print(df_sales.columns)
```

```python
df_sales['total_sales'] = df_sales['sale_price']*df_sales['quantity']
```

```python
df_sales['total_profit'] = df_sales['profit']*df_sales['quantity']
df_sales['profit_margin'] = ((df_sales['sale_price']- df_sales['cost_price'])/df_sales['sale_price'])*100
df_sales
```

```python
df_sales.info()
```

```python
ctor = ['discount','sale_price','profit','total_sales','total_profit']
df_sales[ctor] = df_sales[ctor].round(2)
```

# **Monthly Sales Analysis**



```python
select
extract(month from to_date(order_date, 'DD-MM-YYYY')) as month,
extract(year from to_date(order_date, 'DD-MM-YYYY')) as year,
round(cast(sum(total_sales) as numeric),2) as total_rev,
round(cast(
lag(sum(total_sales)) over (partition by extract(month from to_date(order_date, 'DD-MM-YYYY'))
order by extract(year from to_date(order_date, 'DD-MM-YYYY'))) as numeric),2)as previous_year_revenue,
round(cast(
case
when lag(sum(total_sales)) over (partition by extract(month from to_date(order_date, 'DD-MM-YYYY'))
order by extract(year from to_date(order_date, 'DD-MM-YYYY'))) is not null
then
((sum(total_sales)-lag(sum(total_sales)) over (partition by extract(month from to_date(order_date, 'DD-MM-YYYY'))
order by extract(year from to_date(order_date, 'DD-MM-YYYY')))) / lag(sum(total_sales)) over (partition by extract(month from to_date(order_date, 'DD-MM-YYYY'))
order by extract(year from to_date(order_date, 'DD-MM-YYYY')))) * 100
else null
end as numeric),2) as Year_on_Year_gp
from retail_sales
group by extract(year from to_date(order_date, 'DD-MM-YYYY')),
extract(month from to_date(order_date, 'DD-MM-YYYY'))
order by year desc,month;
```
